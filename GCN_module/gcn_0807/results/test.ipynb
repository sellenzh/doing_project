```python
import matplotlib.pyplot as plt

files = ['train_loss.txt', 'val_loss.txt']
file =  open(files[0], 'r')
data = file.readlines()
loss, itr = [], []
i = 0
for num in data:
    itr.append(i)
    i += 1
    loss.append(float(num))
train_loss_per_half_epoch, train_itr_per = [], []
i, sum = 0, 0
while i < len(loss):
    sum += loss[i]
    i += 1
    if i % 63 == 0:
        sum /= 63
        train_loss_per_half_epoch.append(sum)
        train_itr_per.append(i / 63)

file =  open(files[1], 'r')
data = file.readlines()
loss, itr = [], []
i = 0
for num in data:
    itr.append(i)
    i += 1
    loss.append(float(num))
test_loss_per_half_epoch, test_itr_per = [], []
i, sum = 0, 0
while i < len(loss):
    sum += loss[i]
    i += 1
    if i % 63 == 0:
        sum /= 63
        test_loss_per_half_epoch.append(sum)
        test_itr_per.append(i / 63)


plt.figure()
plt.title('loss')
plt.plot(train_itr_per, train_loss_per_half_epoch, color='red', label='train_loss')
plt.plot(test_itr_per, test_loss_per_half_epoch, color='blue', label='val_loss')
plt.xlabel('x125 iters.')
plt.ylabel('loss_per_mean')
plt.legend()
plt.savefig('loss.jpg')
plt.show()
```
